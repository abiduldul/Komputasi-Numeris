**Bisection** f(x) = x^3 =3x + 1

```python
def f(x):
    return x**3 - 3*x + 1

a = 0
b = 1
epsilon = 1e-5

u = f(a)
v = f(b)

print(f"Initial Interval: [{a}, {b}]")
print("-" * 70)
print("Iter |     a      |     b      |     c      |    f(c)")
print("-" * 70)

iteration = 0
while True:
    iteration += 1

    c = (a + b) / 2
    w = f(c)

    print(f"{iteration:4d} | {a:10.6f} | {b:10.6f} | {c:10.6f} | {w:10.6f}")

    if u * w < 0:
        b = c
        v = w
    else:
        a = c
        u = w

    if (b - a) / 2 < epsilon:
        break

print("-" * 70)
print(f"\nConverged after {iteration} iterations.")
print(f"Approximate root: {c:.6f}")
print(f"Function value at root: {f(c):.6f}")
```
