**Secant Method Code**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return x**5 + x**3 + 3

# Initial values
x0 = -1.0
x1 = -1.1

tol = 0.001
max_iter = 50

print("Iter |     x_i        |     f(x_i)      |   Error")
print("-"*65)

iterations_x = [x0, x1]

for i in range(1, max_iter+1):

    f_x0 = f(x1)
    f_x1 = f(x1)

    # Secant formula
    x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)

    error = abs(x2 - x1)

    print(f"{i:4d} | {x1:12.6f} | {f_x1:14.6f} | {error:10.6f}")

    iterations_x.append(x2)

    if error < tol:
        break

    x0 = x1
    x1 = x2

print("\nConverged Root:")
print(f"x ≈ {x2:.6f}")
print(f"f(x) ≈ {f(x2):.6f}")

# ====== Plot ======
x_vals = np.linspace(-2, 2, 400)
y_vals = f(x_vals)

plt.figure()
plt.plot(x_vals, y_vals)
plt.axhline(0)

# Plot iteration points
for xi in iterations_x:
    plt.scatter(xi, f(xi))

plt.title("Secant Method for f(x) = x^5 + x^3 + 3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
