```python
import numpy as np

# ============================================================
# STEP 1: Input your 3 linear equations below
# Format: [a, b, c, d]  -->  ax + by + cz = d
# ============================================================

eq1 = [ 2,  1, -1,   8]   #  2x +  y -  z =  8
eq2 = [-3, -1,  2, -11]   # -3x -  y + 2z = -11
eq3 = [-2,  1,  2,  -3]   # -2x +  y + 2z = -3

# Build the augmented matrix [A | b]
matrix = np.array([eq1, eq2, eq3], dtype=float)

print('=' * 47)
print('         AUGMENTED MATRIX [A | b]')
print('=' * 47)
print(f"  {'':>4} {'x':>7} {'y':>7} {'z':>7}  {'| d':>7}")
print('-' * 47)
for i, row in enumerate(matrix):
    print(f"  eq{i+1}: {row[0]:>6.2f}  {row[1]:>6.2f}  {row[2]:>6.2f}  | {row[3]:>6.2f}")
print('=' * 47)

# ============================================================
# STEP 2: Gauss-Jordan Elimination (with Partial Pivoting)
# Eliminates BOTH above and below each pivot column
# ============================================================

n = 3
aug = matrix.copy()

print('=' * 47)
print('         GAUSS-JORDAN ELIMINATION')
print('=' * 47)

for col in range(n):
    print(f"\n{'─' * 47}")
    print(f"  >>> Working on pivot column {col+1} (variable {'xyz'[col]})")
    print(f"{'─' * 47}")

    # --- Partial Pivoting ---
    max_row = col + np.argmax(np.abs(aug[col:, col]))
    if max_row != col:
        aug[[col, max_row]] = aug[[max_row, col]]
        print(f"  Swapped row {col+1} <-> row {max_row+1} (partial pivoting)")

    pivot = aug[col, col]
    if pivot == 0:
        raise ValueError(f"Zero pivot at column {col+1}. No unique solution.")

    # --- Normalize pivot row (make pivot = 1) ---
    aug[col] = aug[col] / pivot
    print(f"  R{col+1} <- R{col+1} / {pivot:.4f}   [normalize pivot to 1]")

    # --- Eliminate ALL other rows (above AND below) ---
    for row in range(n):
        if row == col:
            continue
        factor = aug[row, col]
        if factor == 0:
            continue
        aug[row] = aug[row] - factor * aug[col]
        direction = 'below' if row > col else 'above'
        print(f"  R{row+1} <- R{row+1} - ({factor:.4f}) x R{col+1}   [{direction}]")

    print(f"\n  Matrix after column {col+1} elimination:")
    for row in aug:
        print(f"    {row[0]:>8.4f}  {row[1]:>8.4f}  {row[2]:>8.4f}  | {row[3]:>8.4f}")

print('\n' + '=' * 47)
print('       REDUCED ROW ECHELON FORM (RREF)')
print('=' * 47)
print(f"  {'x':>8} {'y':>8} {'z':>8}  {'| d':>8}")
print('-' * 47)
for row in aug:
    print(f"  {row[0]:>8.4f}  {row[1]:>8.4f}  {row[2]:>8.4f}  | {row[3]:>8.4f}")
print('=' * 47)

# ============================================================
# STEP 3: Read Solution Directly
# The augmented column is the solution since [A|b] -> [I|x]
# ============================================================

x = aug[:, n]   # last column is the solution

print('=' * 47)
print('               SOLUTION')
print('  (read directly from the last column)')
print('=' * 47)
print(f"  x = {x[0]:.6f}")
print(f"  y = {x[1]:.6f}")
print(f"  z = {x[2]:.6f}")
print('=' * 47)
```
